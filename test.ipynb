```python
from datetime import datetime, timedelta
```

```python
df_tracon = df.loc[df['airporttraconcode1'].dropna().index]
```

```python
df_tracon['date'] = df_tracon['date'].apply(int).astype(str)
df_tracon = df_tracon.loc[df_tracon['date'].str.len() == 6]
df_tracon['year'] = df_tracon['date'].str.slice(0, 4).astype(int)
df_tracon['month'] = df_tracon['date'].str.slice(4, 6).astype(int)
```

```python
df_tracon_sorted = df_tracon.sort_values(['airporttraconcode1', 'year', 'month'], ascending=['True', 'True', 'True'])
```

```python
df_tracon_sorted_zoa = df_tracon_sorted[(df_tracon_sorted['airporttraconcode1'] == 'ZOA')]
```

```python
df_tracon_sorted_zoa['date'] = pd.to_datetime(df_tracon_sorted_zoa['date'], format='%Y%m')
```

```python
curr_date = df_tracon_sorted_zoa['date'].iloc[0]
```

```python
neg_year_delta = (curr_date.month - 9) // 12
pos_year_delta = (curr_date.month + 9) // 12
```

```python
min_month = (curr_date.month - 9) % 12
max_month = (curr_date.month + 9) % 12
```

```python
min_date = pd.Timestamp(year=curr_date.year + neg_year_delta, month = min_month, day=1)
max_date = pd.Timestamp(year=curr_date.year + pos_year_delta, month = min_month, day=1)
```

```python
min_date <= curr_date <= max_date
```

```python
sliding_window = 6
window = int(sliding_window / 2)
for entry_idx in df_tracon_sorted_zoa.index[5:7]:
    this_index = entry_idx
    
    curr_date = df_tracon_sorted_zoa['date'].loc[entry_idx]
    neg_year_delta, pos_year_delta = ((curr_date.month - 1) - window) // 12, ((curr_date.month - 1) + window) // 12
    min_month, max_month = ((curr_date.month - 1) - window) % 12, ((curr_date.month - 1) + window) % 12
    min_date = pd.Timestamp(year=curr_date.year + neg_year_delta, month = min_month + 1, day=1)
    max_date = pd.Timestamp(year=curr_date.year + pos_year_delta, month = max_month + 1, day=1)
    
    other_df = df_tracon_sorted_zoa[(df_tracon_sorted_zoa['date'] < max_date) & (df_tracon_sorted_zoa['date'] > min_date)]
    other_df = other_df[other_df.index != entry_idx]
    
    print(curr_date)
    print(min_date)
    print(max_date)
    for date in other_df['date']:
          print(date)
    print('\n')
```
